```python
#HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH
keyword = "curso de buceo" # Cambia la keyword por la que quieras (ej. keyword = "descubrimiento de America") 
#HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH
# Par√°metros opcionales
idioma = 'es' # espa√±ol => acepta 'es' (espa√±ol), 'en' (ingl√©s), 'fr' (franc√©s), 'de' (alem√°n), 'it' (italiano), y 'pt' (portugu√©s)
pais = 'es' # Espa√±a => por ejemplo para Inglaterra usar pais = 'uk'
buscarEntidades = True #  False => True busca entidades en la SERP. Tarda unos minutos m√°s pero vale la pena üòâ
#HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH

print("Instalando dependencias...")
!pip install requests
!pip install justext
import requests
import justext
!pip install spacy


# Si quieres a√±adir alg√∫n otro idioma b√∫scalo aqu√≠: https://spacy.io/models#quickstart
if idioma == 'es':
  !python -m spacy download es_core_news_sm
  import es_core_news_sm
  nlp = es_core_news_sm.load()
elif idioma == 'en':
  !python -m spacy download en_core_web_sm
  import en_core_web_sm
  nlp = en_core_web_sm.load()
elif idioma == 'fr':
  !python -m spacy download fr_core_news_sm
  import fr_core_news_sm
  nlp = fr_core_news_sm.load()
elif idioma == 'de':
  !python -m spacy download de_core_news_sm
  import de_core_news_sm
  nlp = de_core_news_sm.load()
elif idioma == 'it':
  !python -m spacy download it_core_news_sm
  import it_core_news_sm
  nlp = it_core_news_sm.load()
elif idioma == 'pt':
  !python -m spacy download pt_core_news_sm
  import pt_core_news_sm
  nlp = pt_core_news_sm.load()


import spacy
from spacy.matcher import Matcher
import requests
from bs4 import BeautifulSoup
! pip install justext
import justext

if buscarEntidades == True:
  # instalamos dependencias
  !pip install wikipedia
  import wikipedia
  wikipedia.set_lang(idioma)
  import random
  import time

print("Scrapeando SERPs...")
class pagina():
    def __init__(self, url):
        self.url = url
        self.texto = ""
        

class serp():
     
    def __init__(self, query):
        self.query = query.replace(" ", "+")
        self.ok = False
        self.incidendias = []
        self.paginas = []
        self.start()

    def start(self):
        URL = f"https://www.google.com/search?hl={idioma}&gl={pais}&q={keyword}&oq={keyword}"
        headers =  {"user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36"}
        enlacesSerp = []
        resp = requests.get(URL, headers = headers)
        if resp.status_code == 200: 
            soup = BeautifulSoup(resp.content, "html.parser")
            links = soup.find_all("div", {"class" : "g"})
            print("Procesando urls:")
            for x in links: #[0:3]
                links0 = x.find_all("a", href=True)
                if len(links0) == 0:continue
                link = links0[0]['href']
                #corrijo feature snippet
                if "#:~:text" in link: 
                    link = link.split("#:~:text")[0]
                if link.startswith("/"):continue
                print(link)

                if not link in enlacesSerp:
                    enlacesSerp.append(link)
                    
                    try:
                      resp = requests.get(link, headers = headers, timeout=10)
                         
                    except Exception as e:
                      self.incidendias.append(f"timeout error: {str(e)} url: {link}")

                    soup = BeautifulSoup(resp.content, "html.parser")

                    if resp.status_code == 200:
                        pag = pagina(link)
                        
                        parrafos = justext.justext(resp.content, justext.get_stoplist("Spanish"))
                        parrafos = [p.text for p in parrafos] #if not p.is_boilerplate
                        
                        pag.texto = '. '.join(parrafos).replace(":.", ":").replace(":,", ",")
                        puntuacion = {".", "?", "¬ø", "!", "¬°"}
                        for p in puntuacion:
                          pag.texto = pag.texto.replace(p + " .", p).replace(p + ".", p).replace(". "+ p, p)

                        self.paginas.append(pag)

                    else:
                        self.incidendias.append(f"Status code: {resp.status_code} url: {link}")

            if len(self.paginas) < 5:
                self.ok = True
                self.incidendias.append("Menos de 5 resultados escrapeados")
            else:
                self.ok = True
 
        else:
            self.ok = False

        #reporto incidencias scrapeo
        print(str(len(self.paginas)) + " p√°ginas escrapeadas correctamente")

        if len(self.incidendias) > 0:
            print("Incidencias")
            print("-----------")
            for x in self.incidendias:
                print(x)

#Aviso para ti, navegante programador => me da pereza limpiar el c√≥digo üòâ
scrap = serp(keyword)

if scrap.ok == False:
    print("No se ha completado el scrap, se detiene el an√°lisis => Sorry...")
    quit()

print("Limpiando texto...")
import re
def limpia(texto):
    frases = re.sub('[^a-zA-Z√Ä-√ø\u00f1\u00d1.¬ø?¬°!]', ' ', texto)
    frases = re.sub(r'\s+', ' ', frases)
    frases = frases.replace("  ", " ").strip()
    frases = frases.replace("(", "").replace(")", "")
    return frases

texto = ""
for pag in scrap.paginas:
  texto = (texto + " " + pag.texto)

texto  = limpia(texto)

print("Creando Nube de palabras...")

def lemmaOTexto(token):
  if token.pos_ == "VERB":
    return token.lemma_
  else:
    return token.text

doc = nlp(texto)
evitar = {"DET", "CONJ", "CCONJ", "ADP", "ADV", "AUX", "PRON", "INTJ", "PUNCT", "SCONJ", "SYM"}

soloPalabras = [lemmaOTexto(token).lower()
        for token in doc
        if not token.is_stop and not token.is_punct and token.pos_ not in evitar]

from wordcloud import WordCloud
import matplotlib.pyplot as plt 
wordcloud = WordCloud(background_color='white',
                            max_words=100,
                            max_font_size=50, 
                            random_state=42
                            ).generate(str(soloPalabras)) #, collocations=False
plt.figure(figsize=(13,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()

#Visualizaci√≥n de tablas
import pandas as pd
from google.colab import data_table
def pasarATabla(lista,columnas):
  lista0 = pd.DataFrame (lista, columns = columnas )
  lista1 = data_table.DataTable(lista0, include_index=True, num_rows_per_page=20)
  display(lista1)
  return lista0

print("----------------------------------------------------------------------------")
print("He visto cosas que jamas creer√≠ais:")
print("=> SERPs sin anuncios ni featured snippets ...")
print("=> Granjas de enlaces gratuitos que pasaban autorizadad m√°s alla de ori√≥n...")
print("Todos esos recuerdos se perder√°n en el tiempo como l√°grimas en la lluvia üòÖ")
print("----------------------------------------------------------------------------")

print("Creando tabla por posici√≥n gramatical (sustantivo, verbo, adjetivo...)")
from collections import Counter
words3 = [lemmaOTexto(token).lower() + "__" + token.pos_
    for token in doc
        if not token.is_stop and not token.is_punct and token.pos_ not in evitar]

word_freq = Counter(words3)
common_words = word_freq.most_common()

words3 = [[x[0].split("__")[0], x[0].split("__")[1],  x[1] ] for x in list(common_words)]

#agrupo duplicados que caigan en otra categor√≠a
wordsAgrupadas = []
for palabra in words3:
  if palabra[0] in [x[0] for x in wordsAgrupadas]: continue
  espejo = filter(lambda item: item[0] == palabra[0], words3)
  num = sum([x[2] for x in espejo])
  wordsAgrupadas.append([palabra[0], palabra[1], num])

#reordeno
wordsAgrupadas.sort(key=lambda x: x[2], reverse=True)

tablaAdjetivos = pasarATabla(wordsAgrupadas, ["Palabra", "Tipo", "Frecuencia"])


# Adjetivo + sustantivo
nouns = [token.text
    for token in doc
    if (not token.is_stop and
        not token.is_punct and
        token.pos_ == "NOUN" or token.pos_ == "PROPN" )] 

noun_freq = Counter(nouns)
sustantivos = noun_freq.most_common(30)

matcher = Matcher(nlp.vocab)
for masComun in sustantivos: #[:20]
  sustantivo = str(masComun[0])
  pattern = [
    [{"POS": "ADJ"}, {"TEXT": sustantivo}],
    [{"POS": "ADJ"}, {"TEXT": sustantivo.capitalize()}],
    [{"TEXT": sustantivo}, {"POS": "ADJ"} ],
  [{"TEXT": sustantivo.capitalize()}, {"POS": "ADJ"} ]
  ]
  matcher.add("ADJ_NOUN_PATTERN", pattern)
  matches = matcher(doc)


relevantes = []
for match_id, start, end in matches:
  relevantes.append(doc[start:end].text.lower())


relevantes = Counter(relevantes)
pasarATabla(relevantes.most_common(), ["Adjetivo + Sustantivo", "Frecuencia"])

#Busco agrupaciones de palabras m√°s frecuentes
from nltk.tag import pos_tag
from collections import Counter
from nltk import ngrams

def devuelveNgrams(texto, n):
  xgrams = list(ngrams(texto.split(), n))
  xgrams = [" ".join(gram) for gram in xgrams]
  word_freq = Counter(xgrams)
  return word_freq

soloPalabras0 = [token.text.lower()
        for token in doc
        if not token.is_stop and not token.is_punct and token.pos_ not in evitar]

entero = " ".join(soloPalabras0)
bigrams = devuelveNgrams(entero, 2).most_common(10)
trigrams = devuelveNgrams(entero, 3).most_common(10)

import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas

sns.set()

def mostrarGrafica(lista, titulo, columnas):
  top_df = pandas.DataFrame(lista)
  top_df.columns=["Palabras", "Frecuencia"]
  sns.set(rc={'figure.figsize':(13,8)})
  g = sns.barplot(x=columnas[0], y=columnas[1], data=top_df)
  g.axes.set_title(titulo,fontsize=25)
  g.set_xticklabels(g.get_xticklabels(), rotation=50)
  plt.show()

mostrarGrafica(bigrams, "Agrupaciones de 2 palabras", ["Palabras", "Frecuencia"])
mostrarGrafica(trigrams, "Agrupaciones de 3 palabras", ["Palabras", "Frecuencia"])

# ¬øbuscamos entidades?
if buscarEntidades == True:
  print("--------------------------------------------------------------------------------------------------")
  print("Buscando entidades, este proceso puede llevar unos 3 o 5 minutos dependiendo de la API de wikipedia...")
  print("--------------------------------------------------------------------------------------------------")
  # para entidades de 2 o m√°s palabras quito stopwords (de, la, que...)
  stopwords = list(nlp.Defaults.stop_words)
  
  # funci√≥n principal de b√∫squeda de entidades con API de wikipedia
  def devuelveEntidades(numPalabras, minimoContador):
      entidades = {}
      grams = devuelveNgrams(entero, numPalabras).most_common()
      grams = [x for x in grams if x[1] >= minimoContador]
      for x in grams[:250]:
        if len(x[0]) < 2: continue
        
        try:
          # la api de wikipedia nos devuelve las entidades/p√°ginas relacionadas
          salidaWiki = wikipedia.search(x[0])
          salida = [z for z in salidaWiki] 
          # comprobamos si la entidad exacta existe
          if numPalabras == 1: #para 1 palabra voy a tiro hecho, debe coincidir
            salida2 = [x.lower() for x in salida]
            palabra = x[0].lower()
            if palabra in salida2:
              entidad = salida[salida2.index(palabra)]
              if entidad in entidades.keys():
                entidades[entidad] = entidades[entidad] + x[1]
              else:
                entidades[entidad] = x[1]
          else: # Para m√°s de una palabra quito stopwords y comparo listas => ej edad de bronce = ["edad", "bronce"]
            salida2 = [[y for y in z.lower().split() if y not in stopwords] for z in salida]
            palabraSplit = [y for y in x[0].lower().split() if y not in stopwords]
            if palabraSplit in salida2:
              entidad = salida[salida2.index(palabraSplit)]
              if entidad in entidades.keys():
                entidades[entidad] = entidades[entidad] + x[1]
              else:
                entidades[entidad] = x[1]
        except:
          True #Nada que decir, a veces falla...

      # Paso el listado a tabla 
      listado = list(entidades.items())
      if len(listado) > 0:
        pasarATabla(listado, [f"Entidades - {numPalabras} palabra(s) principal(es)", "Contador"])

      comienza = random.randint(5, 10)/1000 # pongamos una peque√±a espera random
      time.sleep(comienza)

      return # entidades

  # Buscamos entidades para 3 palabras principales, 2 palabras, 1 palabra
  devuelveEntidades(numPalabras = 3, minimoContador = 3)
  devuelveEntidades(numPalabras = 2, minimoContador = 4)  
  devuelveEntidades(numPalabras = 1, minimoContador = 15)
```
